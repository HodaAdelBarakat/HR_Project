```python
# Merge the first name and last name rows into the full name row and delete the first two rows
employee_df["FullName"] = employee_df["FirstName"] + " " + employee_df["LastName"]
employee_df.drop(["FirstName", "LastName"], axis=1, inplace=True)
```

```python
# Combine first and last name into FullName column
employee_df["FullName"] = employee_df["FirstName"] + " " + employee_df["LastName"]
employee_df.drop(["FirstName", "LastName"], axis=1, inplace=True)
cols = employee_df.columns.tolist()
cols.insert(1, "FullName")
employee_df = employee_df.reindex(columns=cols)
```

```python
# Replace education codes with labels
education_df = education_df.rename(columns={"EducationLevelID" : "Education"})
education_map = dict(zip(education_df["Education"], education_df["EducationLevel"]))
employee_df["Education"] = employee_df["Education"].map(education_map)
```

```python
# Map satisfaction levels
satisfaction_map = dict(zip(satisfaction_df["SatisfactionID"], satisfaction_df["SatisfactionLevel"]))
performance_df["EnvironmentSatisfaction"] = performance_df["EnvironmentSatisfaction"].map(satisfaction_map)
performance_df["JobSatisfaction"] = performance_df["JobSatisfaction"].map(satisfaction_map)
performance_df["RelationshipSatisfaction"] = performance_df["RelationshipSatisfaction"].map(satisfaction_map)
```

```python
# Map rating levels
rating_map = dict(zip(rating_df["RatingID"], rating_df["RatingLevel"]))
performance_df["SelfRating"] = performance_df["SelfRating"].map(rating_map)
performance_df["ManagerRating"] = performance_df["ManagerRating"].map(rating_map)
```

```python
merged_df = employee_df.merge(performance_df, on="EmployeeID", how="left")
```

```python
#Q4. How will the average salary in the company change over the next year?
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = employee_df[['Age', 'YearsAtCompany']]
y = employee_df['Salary']
model = LinearRegression().fit(X, y)

future_data = X.copy()
future_data['YearsAtCompany'] += 1
employee_df['PredictedSalaryNextYear'] = model.predict(future_data)

current_avg_salary = employee_df['Salary'].mean()
predicted_avg_salary = employee_df['PredictedSalaryNextYear'].mean()

print(f"Current Average Salary: ${current_avg_salary:.2f}")
print(f"Predicted Average Salary Next Year: ${predicted_avg_salary:.2f}")
```

```python
# Q5: How many employees are expected to be hired in each department next year?
#  HireDate  datetime    
employee_df['HireDate'] = pd.to_datetime(employee_df['HireDate'])

#   
employee_df['HireYear'] = employee_df['HireDate'].dt.year
#        
hire_trend = employee_df.groupby(['HireYear', 'Department']).size().reset_index(name='Hires')

#  
plt.figure(figsize=(12, 6))
sns.lineplot(data=hire_trend, x='HireYear', y='Hires', hue='Department', marker='o')
plt.title('Hiring Trend by Department')
plt.ylabel('Number of Hires')
plt.xlabel('Year')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Q5: How many employees are expected to be hired in each department next year?
#         Linear Regression  

from sklearn.linear_model import LinearRegression

future_hiring = {}

for dept in hire_trend['Department'].unique():
    dept_data = hire_trend[hire_trend['Department'] == dept]
    X = dept_data[['HireYear']]
    y = dept_data['Hires']

    model = LinearRegression().fit(X, y)
    next_year = [[employee_df['HireYear'].max() + 1]]
    prediction = model.predict(next_year)[0]

    future_hiring[dept] = int(round(prediction))

print("Expected hires next year per department:")
for dept, hires in future_hiring.items():
    print(f"{dept}: {hires} hires")
```

```python
# Q7: How do work-life balance and frequent travel impact employee satisfaction and retention?
#Filter relevant columns
df_q8 = merged_df[['BusinessTravel', 'WorkLifeBalance', 'JobSatisfaction', 'Attrition']]

sns.set_theme(style="whitegrid")

# Create two subplots: One for satisfaction, one for attrition
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# --- Plot 1: Work-Life Balance vs Job Satisfaction ---
sns.boxplot(x='WorkLifeBalance',
            y='JobSatisfaction',
            data=df_q8,
            palette="coolwarm",
            ax=axes[0])
axes[0].set_title('Job Satisfaction vs Work-Life Balance', fontsize=13)
axes[0].set_xlabel('Work-Life Balance (1 = Bad, 5 = Excellent)')
axes[0].set_ylabel('Job Satisfaction')

# --- Plot 2: Business Travel vs Attrition ---
sns.countplot(x='BusinessTravel',
              hue='Attrition',
              data=df_q8,
              palette="Set2",
              ax=axes[1])
axes[1].set_title('Attrition vs Business Travel', fontsize=13)
axes[1].set_xlabel('Business Travel Frequency')
axes[1].set_ylabel('Number of Employees')
axes[1].tick_params(axis='x', rotation=15)

plt.tight_layout()
plt.show()
```

```python
# Question 8: How does the number of years with the same manager affect employee satisfaction and retention?

sns.set_theme(style="whitegrid")

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# --- Plot 1: Boxplot - Years with Manager vs Job Satisfaction ---
sns.boxplot(x='JobSatisfaction', y='YearsWithCurrManager',
            data=merged_df, palette="pastel", ax=axes[0])
axes[0].set_title('Years with Current Manager vs Job Satisfaction', fontsize=13)
axes[0].set_xlabel('Job Satisfaction')
axes[0].set_ylabel('Years with Current Manager')

# --- Plot 2: Stacked Histogram - Attrition vs Years with Manager ---
sns.histplot(data=merged_df,
             x='YearsWithCurrManager',
             hue='Attrition',
             multiple='stack',
             bins=15,
             palette='Set2',
             ax=axes[1])
axes[1].set_title('Attrition vs Years with Current Manager', fontsize=13)
axes[1].set_xlabel('Years with Manager')
axes[1].set_ylabel('Number of Employees')

plt.tight_layout()
plt.show()
```

```python
#Question 9: How realistic is the gap between employees' self-ratings and manager ratings?
rating_comparison = performance_df[['SelfRating', 'ManagerRating']].copy()

comparison_table = pd.crosstab(rating_comparison['SelfRating'],
                               rating_comparison['ManagerRating'])

print("Self vs Manager Rating Comparison Table:")
print(comparison_table)

plt.figure(figsize=(10,6))
sns.heatmap(comparison_table, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Self Rating vs Manager Rating")
plt.xlabel("Manager Rating")
plt.ylabel("Self Rating")
plt.tight_layout()
plt.show()
```
